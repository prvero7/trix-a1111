```python
#@markdown # Model Download/Load

import gdown
from gdown.download import get_url_from_gdrive_confirmation
import re
from google.colab import files
import json

Use_Temp_Storage = False #@param {type:"boolean"}
#@markdown - If not, make sure you have enough space on your gdrive

#@markdown ---

Use_Default_Models = True #@param {type:"boolean"}
#@markdown - If you'd like download custom models, create a schema following example [here](https://github.com/prvero7/trix-a1111/blob/main/models-schema.json)

DEFAULT_SCHEMAS="https://raw.githubusercontent.com/prvero7/trix-a1111/refs/heads/main/models-schema.json"

def get_src_from_url(url):
    parsed_url = urlparse(url)
    if parsed_url.netloc == 'civitai.com':
        src='civitai'
    elif parsed_url.netloc == 'drive.google.com':
        src='gdrive'
    elif parsed_url.netloc == 'huggingface.co':
        src='huggingface'
    else:
        src='others'
    return src

def auth_request(url):
  src = get_src_from_url(url)
  headers = {"User-Agent": "torch.hub"}

  if src=='civitai':
    if url.count("?") > 0:
      url += f'&token={CIVIT_API}'
    else:
      url += f'?token={CIVIT_API}'

  if src=='huggingface':
    headers["Authorization"] = f"Bearer {HUGGINGFACE_API}"

  r=Request(url, headers=headers)
  return r

def get_name(url, gdrive):
    if not gdrive:
        response = requests.get(url, allow_redirects=False)
        if "Location" in response.headers:
            redirected_url = response.headers["Location"]
            quer = parse_qs(urlparse(redirected_url).query)
            if "response-content-disposition" in quer:
                disp_val = quer["response-content-disposition"][0].split(";")
                for vals in disp_val:
                    if vals.strip().startswith("filename="):
                        filenm=unquote(vals.split("=", 1)[1].strip())
                        return filenm.replace("\"","")
        return url.split("/")[-1].split("?")[0]
    else:
        headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36"}
        lnk="https://drive.google.com/uc?id={id}&export=download".format(id=url[url.find("/d/")+3:url.find("/view")])
        res = requests.session().get(lnk, headers=headers, stream=True, verify=True)
        res = requests.session().get(get_url_from_gdrive_confirmation(res.text), headers=headers, stream=True, verify=True)
        content_disposition = six.moves.urllib_parse.unquote(res.headers["Content-Disposition"])
        filenm = re.search('attachment; filename="(.*?)"', content_disposition).groups()[0]
        return filenm


def dwn(url, dst, msg):
    file_size = None
    req = auth_request(url)
    u = urlopen(req)
    meta = u.info()
    if hasattr(meta, 'getheaders'):
        content_length = meta.getheaders("Content-Length")
    else:
        content_length = meta.get_all("Content-Length")
    if content_length is not None and len(content_length) > 0:
        file_size = int(content_length[0])

    with tqdm(total=file_size, disable=False, mininterval=0.5,
              bar_format=msg+' |{bar:20}| {percentage:3.0f}%') as pbar:
        with open(dst, "wb") as f:
            while True:
                buffer = u.read(8192)
                if len(buffer) == 0:
                    break
                f.write(buffer)
                pbar.update(len(buffer))
            f.close()


def read_schema_file(filepath) -> list[dict[str,str]]:
  with open(filepath, 'r') as file:
      file_contents = file.read()

  return json.loads(file_contents)


def download_models_from_schema(schema_path, Use_Temp_Storage):
  schemas = read_schema_file(schema_path)

  for schema in schemas:
    link=schema["url"]
    if Use_Temp_Storage:
        os.makedirs('/content/temp_models', exist_ok=True)
        model=f'/content/temp_models/{schema["name"]}'
    else:
      model='/content/gdrive/'+mainpth+'/sd/stable-diffusion-w'+blsaphemy+f'/models/Stable-diffusion/{schema["name"]}'

    if not os.path.exists(model):
      dwn(link, model, f'Downloading the model {schema["name"]}')
    else:
      print(f'Model {schema["name"]} already exists')


clear_output()
if Use_Default_Models:
    schema_path = "/content/models-schema.json"
    !wget -q -O $schema_path -i https://raw.githubusercontent.com/prvero7/trix-a1111/refs/heads/main/models-schema.json

    download_models_from_schema(schema_path, Use_Temp_Storage)

    !rm $schema_path
else:
  filepath = files.upload(target_dir="/content")
  filepath = list(filepath.keys())[0]

  download_models_from_schema(filepath, Use_Temp_Storage)

  !rm -rf $filepath

#@markdown ---
```
